```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
  
chi2 = np.load('../chi2_2mode_time_avg.npz')['chi2_time_avg']
ax = sns.heatmap( chi2 , linewidth = 0.5 , cmap = 'coolwarm' )
ax.tick_params(axis="x", rotation=90)
ax.tick_params(axis="y", rotation=0)
plt.title(r'$\langle\chi^2\rangle_t$, 2-mode combinations')
#plt.xlabel('mode')
#plt.ylabel('mode')
fname = 'chi2_2mode_time_avg.png'
plt.savefig(fname, dpi=450, facecolor='w', 
        edgecolor='w', orientation='portrait', 
        format='png', transparent=False, 
        bbox_inches='tight', pad_inches=0.1, 
        metadata=None)

# colour gradients: 
# 'PiYG', 'coolwarm'
# ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
# 'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
# 'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'])
```
