```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

# load reference data
datafile = '../data/NMM_exp_dataset.mat'
mat = scipy.io.loadmat(datafile)
q = np.squeeze(mat['q'])
t = np.squeeze(mat['t'])
print(t)
print('number of timesteps: %i' % len(t))
print(len(q))
print(q)
data = mat['iso']
errors = mat['iso_stdx']

plt.figure()
plt.pcolormesh(t, q, data)
plt.xlabel('t (ps)')
plt.ylabel('q ($\AA^{-1}$)')
plt.title('$\%\Delta I(q,t)$')
plt.colorbar()
fname = 'NMM_exp_2Dplot.png'
dpi = 300
plt.savefig(fname, dpi=dpi, facecolor='w', 
            edgecolor='w', orientation='portrait', 
            format='png', transparent=False, 
            bbox_inches='tight', pad_inches=0.1, 
            metadata=None)
```
